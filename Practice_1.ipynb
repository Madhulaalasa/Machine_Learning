```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
d=pd.read_csv("canada_per_capita_income.csv")
d.head(2)
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Capita')
plt.scatter(d.year,d.capita)
```

```python
reg=linear_model.LinearRegression()
reg.fit(d[['year']].values,d.capita)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Capita')
plt.scatter(d.year,d.capita)
plt.plot(d.year,reg.predict(d[["year"]].values))
```
